```python
# ---
# Audio Agent Demo Notebook
# ---
import os, sys
# Add project root to sys.path
sys.path.append(os.path.abspath(".."))


import os, sys, requests
from IPython.display import Audio, display
from app.utils import convert_ogg_to_wav  
# ---
# Convert OGG â†’ WAV (if needed)
input_path = "../data/sample.ogg"
wav_path = convert_ogg_to_wav(input_path)
display(Audio(wav_path))

# ---
#  Send the WAV file to the running FastAPI endpoint
API_URL = "http://localhost:8000/process_audio"

with open(wav_path, "rb") as f:
    response = requests.post(API_URL, files={"file": f})

# ---
# Display API response
if response.status_code == 200:
    result = response.json()
    print("Transcription and agent response received!\n")
    print("Original Text:")
    print(result["original"])
    print("\nCorrected Text:")
    print(result["corrected"])
    print("\nAgent Response:")
    print(result["agent_response"])
else:
    print("API Error:", response.status_code, response.text)

# ---
# Optional: test multiple files automatically
# from glob import glob
# for ogg in glob("../data/*.ogg"):
#     wav = convert_ogg_to_wav(ogg)
#     with open(wav, "rb") as f:
#         r = requests.post(API_URL, files={"file": f})
#         print(ogg, "â†’", r.json().get("corrected"))
```
